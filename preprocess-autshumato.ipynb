```python
import os
import chardet
import codecs
import re

import numpy as np
import pandas as pd
```

```python
ISO_MAP = {
    'Afrikaans': 'af',
    'English': 'en',
    'IsiNdebele': 'nr',
    'IsiZulu': 'zu',
    'IsiXhosa': 'xh',
    'Sepedi': 'nso',
    'Sesotho': 'st',
    'Setswana': 'tn',
    'Siswati': 'ss',
    'Tshivenda': 've',
    'Xitsonga': 'ts'
}
```

```python
data_dir = './data/raw'
proc_dir = './data/processed'
os.listdir(data_dir)
```

```python
def process(line):
    line = line.strip()
    line = line + '\n'
    if re.match(r"(<Doc[0-9]*>\n)", line):
        return False, line
    return True, line
```

```python
for fp in os.listdir(data_dir):
    _, _, lang, translator, fmt = fp.split('.')
    with open(os.path.join(data_dir, fp), 'rb') as text:
        char_enc = chardet.detect(text.read())
    with codecs.open(os.path.join(data_dir, fp), 'r', char_enc['encoding']) as text:
        lines = text.readlines()
    
    proc_name = f"{translator.lower()}.{ISO_MAP[lang]}.txt"
    
    print(proc_name)
    print('-'*len(proc_name), end='\n')
    
    proc = codecs.open(os.path.join(proc_dir, proc_name), 'w', 'utf-8')
    wrote = 0
    for line in lines:
        write, processed = process(line)
        if write:
            proc.write(processed)
            wrote+=1
        else:
            print(f"Caught bad line {processed}", end="\r")
    print(f'Wrote {wrote} lines to {proc_name}.\n')
    proc.close()
#     proc = open(os.path.join(proc_dir, translator.lower() + '.' + ISO_MAP[lang]), 'w')
    
        
```
